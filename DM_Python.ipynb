### **INTRODUCTION**
---
> Les marchés des télécommunications sont parmi les plus concurrentiels au monde, avec de nombreux fournisseurs dans chaque pays. Pour rester compétitives, ces entreprises se concentrent sur la gestion de la relation client et la fidélisation, car acquérir un nouveau client peut coûter jusqu'à 700 % de plus que de conserver un client existant.
```python
from scipy.stats import chi2_contingency

# Liste de toutes les variables catégorielles que vous souhaitez tester
categorical_columns = df.select_dtypes(include=['object'])  # Sélectionnez les colonnes catégorielles

# Variable cible
target_variable = 'Churn'

# Seuil de signification
seuil = 0.05

# Créez un dictionnaire pour stocker les résultats
resultats_chi2 = {}
variables_corrélées = []
variables_non_corrélées = []

# Effectuez des tests du chi-carré pour chaque paire de variables
for i in range(len(categorical_columns.columns)):
    variable1 = df[categorical_columns.columns[i]]

    # Créez un tableau de contingence
    contingency_table = pd.crosstab(variable1, df[target_variable])

    # Appliquez le test du chi-carré
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Stockez les résultats dans le dictionnaire
    nom_resultat = f"{categorical_columns.columns[i]} vs {target_variable}"
    resultats_chi2[nom_resultat] = {'Statistique du chi-carré': chi2, 'Valeur p': p, 'Degrés de liberté': dof}

    # Vérifiez si la valeur p est inférieure au seuil
    if p < seuil:
        variables_corrélées.append(categorical_columns.columns[i])
    else:
        variables_non_corrélées.append(categorical_columns.columns[i])

# Affichez les résultats
for nom_resultat, resultat in resultats_chi2.items():
    print(nom_resultat)
    for mesure, valeur in resultat.items():
        print(f"{mesure}: {valeur}")
    print("\n")

# Affichez les variables catégorielles corrélées à la variable cible 'Churn'
print("Variables catégorielles corrélées à la variable cible 'Churn':")
print(variables_corrélées)

# Affichez les variables catégorielles non corrélées à la variable cible 'Churn'
print("Variables catégorielles non corrélées à la variable cible 'Churn':")
print(variables_non_corrélées)
```

**b- treatement of  skewed variables**

---



 on utiliser des mesures statistiques telles que la skewness pour évaluer quantitativement l'asymétrie de la distribution des données. La skewness est une mesure numérique qui indique l'asymétrie de la distribution des données. Une valeur de skewness positive indique une distribution à droite (right-skew), tandis qu'une valeur négative indique une distribution à gauche (left-skew). Une valeur proche de zéro suggère une distribution relativement symétrique."

```python
# Create an empty list to store the names of skewed variables
skewed_variables = []

# Create a DataFrame to store the log-transformed variables
log_transformed_df = pd.DataFrame()

# Set a threshold for skewness (e.g., 1.0 or -1.0) based on your criteria
skewness_threshold = 1.0

# Iterate through columns in your DataFrame
for column in df.columns:
    # Check if the column is numeric (you can adjust the data type check as needed)
    if pd.api.types.is_numeric_dtype(df[column]):
        # Calculate the skewness measure
        skewness = df[column].skew()

        # Check if the skewness exceeds the threshold
        if abs(skewness) > skewness_threshold:
            print(f"Variable '{column}' is skewed (Skewness = {skewness:.4f})")

            # Apply a log transformation to the skewed variable and store it in the log_transformed_df
            log_transformed_df[column] = df[column].apply(lambda x: x if x <= 0 else np.log(x))

            # Add the variable name to the list of skewed variables
            skewed_variables.append(column)
        else:
            # If not skewed, add the original column to the log_transformed_df
            log_transformed_df[column] = df[column]
    else:
        # If the column is not numeric, add it to the log_transformed_df as is
        log_transformed_df[column] = df[column]

# Print the list of skewed variables
print("Skewed Variables:", skewed_variables)
```

All variables are skewed that's why we applied the log function to those skewed variables

**c- Dealing with Outliers**

---



**Au lieu de les supprimer et de perdre des données précieuses, les valeurs aberrantes sont traitées en les remplaçant par des valeurs médianes. Toutes les valeurs au-dessus du 99e percentile sont remplacées par la valeur médiane. Cette technique est appliquée uniquement aux caractéristiques continues.**

```python
# List of symetric variables to exclude
exclude_variables = ['AgeHH1', 'AgeHH2', 'IncomeGroup']

# Define the percentile threshold (e.g., 99th percentile)
percentile_threshold = 99

# Create a new list of numeric variables, excluding the ones in 'exclude_variables'
numeric_variables = [var for var in continuous_vars if var not in exclude_variables]

# Iterate through each numeric variable
for var_name in numeric_variables:
    # Get the data for the current variable
    current_variable = log_transformed_df[var_name]

    # Identify the upper threshold value based on the percentile
    upper_threshold = np.percentile(current_variable, percentile_threshold)

    # Replace values above the upper threshold with the median
    current_variable[current_variable > upper_threshold] = np.median(current_variable)
```
